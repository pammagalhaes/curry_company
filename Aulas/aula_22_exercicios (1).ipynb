````python
lista = []

# range( inicio, fim, [quantidade] ) | range(fim)
for valor in list(range(51)): # [ 0:51 [
  print(valor)
  print('Um valor foi removido da lista')
  lista.append(valor)

  print()

print('Quantidade de elementos na lista')
len(lista)  # length
````
